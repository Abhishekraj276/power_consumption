```python
df["Timestamp"]=pd.to_datetime(df['Timestamp'])
df=df.sort_values('Timestamp')
```

```python
df = df.set_index('Timestamp')
```

```python
df.head()
```

```python
df['HVACUsage'] = df['HVACUsage'].map({'On': 1, 'Off': 0})
df['LightingUsage'] = df['LightingUsage'].map({'On': 1, 'Off': 0})
df['Holiday'] = df['Holiday'].map({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
day_map = {'Monday': 0, 'Tuesday': 1, 'Wednesday': 2, 'Thursday': 3, 
           'Friday': 4, 'Saturday': 5, 'Sunday': 6}
df['DayOfWeek'] = df['DayOfWeek'].map(day_map)
```

```python
df.head()
```

```python
df['Hour'] = df.index.hour
df['DayOfMonth'] = df.index.day
df['Month'] = df.index.month
df['WeekOfYear'] = df.index.isocalendar().week
df['IsWeekend'] = (df.index.dayofweek >= 5).astype(int)
df['Season'] = (df.index.month % 12 + 3) // 3  
```

```python
df['Energy_Lag_24h'] = df['EnergyConsumption'].shift(24)
df['Energy_Lag_168h'] = df['EnergyConsumption'].shift(168) 
df['Energy_MA_24h'] = df['EnergyConsumption'].rolling(window=24).mean()
df['Energy_MA_7d'] = df['EnergyConsumption'].rolling(window=168).mean()  
df = df.dropna()
```

```python
split_index = int(len(df) * 0.8)
train = df.iloc[:split_index]
test = df.iloc[split_index:]

print(f"Train size: {len(train)}")
print(f"Test size: {len(test)}")
```

```python
from sklearn.model_selection import TimeSeriesSplit

tscv = TimeSeriesSplit(n_splits=5)
for train_index, test_index in tscv.split(df):
    train, test = df.iloc[train_index], df.iloc[test_index]
    print(f"Train: {len(train)}, Test: {len(test)}")
```

```python
from sklearn.preprocessing import StandardScaler

X_train = train.drop('EnergyConsumption', axis=1)
y_train = train['EnergyConsumption']
X_test = test.drop('EnergyConsumption', axis=1)
y_test = test['EnergyConsumption']


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,root_mean_squared_error

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print(f"MAE: {mean_absolute_error(y_test, y_pred):.2f}")
print(f"RMSE: {root_mean_squared_error(y_test, y_pred):.2f}")
print(f"R: {r2_score(y_test, y_pred):.3f}")
```

```python
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score

models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42),
    'XGBoost': XGBRegressor(random_state=42, verbosity=0)
}

results =[]

for name,model in models.items():

    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)

    mae = mean_absolute_error(y_test, y_pred)
    rmse = root_mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results.append({
        'Model': name,
        'MAE': mae,
        'RMSE': rmse,
        'R': r2
    })
    
    print(f"{name}: MAE={mae:.2f}, RMSE={rmse:.2f}, R={r2:.3f}")
```

```python
results_df = pd.DataFrame(results)
print("\nModel Comparison:")
print(results_df.sort_values('MAE'))
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error

train_energy = y_train.values
test_energy = y_test.values

try:
    arima_model = ARIMA(train_energy, order=(1,1,1))
    arima_result = arima_model.fit()

    arima_forecast = arima_result.forecast(steps=len(test_energy))
    arima_mae = mean_absolute_error(test_energy, arima_forecast)
    arima_rmse = root_mean_squared_error(test_energy, arima_forecast)
    arima_r2 = r2_score(test_energy, arima_forecast)
    print(f"ARIMA: MAE={arima_mae:.2f}, RMSE={arima_rmse:.2f}, R={arima_r2:.3f}")
    results_df = pd.concat([results_df, pd.DataFrame([{
        'Model': 'ARIMA',
        'MAE': arima_mae,
        'RMSE': arima_rmse,
        'R': arima_r2
    }])], ignore_index=True)

except Exception as e:
    print(f"ARIMA failed: {e}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.bar(results_df['Model'], results_df['MAE'])
plt.title('MAE Comparison')
plt.xticks(rotation=45)

plt.subplot(2, 2, 2)
plt.bar(results_df['Model'], results_df['RMSE'])
plt.title('RMSE Comparison')
plt.xticks(rotation=45)

plt.subplot(2, 2, 3)
plt.bar(results_df['Model'], results_df['R'])
plt.title('R Comparison')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

print("Models sorted by MAE (best first):")
print(results_df.sort_values('MAE')[['Model', 'MAE', 'RMSE', 'R']])
```
